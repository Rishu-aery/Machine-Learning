```python
#Load the Libraries
from sklearn import datasets
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from matplotlib import pyplot as plt
```

```python
#Load data with only two features
boston=datasets.load_boston()
features=boston.data[:,0:2]
target=boston.target

```

```python
# Create decision tree regressor object
regr_tree=DecisionTreeRegressor(max_depth=2,random_state=0)
# Train model
model=regr_tree.fit(features,target)
```

```python
# Visualize the tree
flg=plt.figure(figsize=(11,8))
_=tree.plot_tree(regr_tree,feature_names=boston.feature_names,filled=True)
```

```python
# Make new observation
observation =[[0.02,16]]
#predict observations value
model.predict(observation)
```

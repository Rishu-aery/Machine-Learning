```python
#Load the Libraries
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot as plt
```

```python
#Load data and fit the model
iris=load_iris()
x=iris.data
y=iris.target
tree_clf=DecisionTreeClassifier(max_depth=2)
tree_clf.fit(x,y)
```

```python
#Visualize the tree
flg=plt.figure(figsize=(8,8))
_=tree.plot_tree(tree_clf,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
```

```python
tree_clf.predict_proba([[2,3,5,1.5]])
```
